### Padronizando unidades de medida

```python
def standardize_units_usd_kg(df, densities):  # Padronização para dólares por quilograma
    conversion_factors = {
        'Preço lã grossa': 0.01,  # Centavos de dólar para dólar
        'Preço copra': 1/1000,  # Tonelada para kg
        'Preço algodão': 1,  # No pradão
        'Preço lã fina': 0.01,  # Centavos de dólar para dólar
        'Preço tora dura': 1/densities['Preço tora dura'],  # Metro cúbico para kg
        'Preço madeira serrada dura': 1/densities['Preço madeira serrada dura'],  # Metro cúbico para kg
        'Preço couro': 0.01 / 0.453592,  # Centavos de dólar por libra para dólar por kg
        'Preço madeira compensada': 1/densities['Preço madeira compensada'],  # Metro cúbico para kg
        'Preço borracha': 1,  # No pradão
        'Preço Tora macia': 1/densities['Preço Tora macia'],  # Metro cúbico para kg
        'Preço madeira serrada macia': 1/densities['Preço madeira serrada macia'],  # Metro cúbico para kg
        'Preço celulose': 1/1000  # Tonelada para kg
    }

    df_converted = df.copy()
    for column, factor in conversion_factors.items():
        df_converted[column] = df[column] * factor

    return df_converted

densities = {
    'Preço tora dura': 700,  
    'Preço madeira serrada dura': 800,  
    'Preço madeira compensada': 600,  
    'Preço Tora macia': 500, 
    'Preço madeira serrada macia': 400 
}

df = standardize_units_usd_kg(df, densities)
```

### **Matriz de correlação**

**CONTEXTO**

Uma matriz de correlação é uma ferramenta estatística que apresenta a força e a direção da relação entre duas ou mais variáveis em um conjunto de dados. Ela é representada por uma tabela quadrada na qual cada célula indica o coeficiente de correlação entre um par específico de variáveis.

O coeficiente de correlação varia entre -1 e 1, indicando a intensidade e a direção da relação entre as variáveis:

* **Valor positivo:** indica uma correlação positiva, ou seja, as variáveis tendem a aumentar ou diminuir juntas. Quanto mais próximo de 1, mais forte a correlação.
* **Valor negativo:** indica uma correlação negativa, ou seja, as variáveis tendem a se mover em direções opostas. Quanto mais próximo de -1, mais forte a correlação.
* **Valor 0 (zero):** indica que não há correlação linear entre as variáveis, ou seja, seus movimentos não estão relacionados entre si.

```python
def plot_heatmap(corr_matrix):
    fig = plt.figure(figsize = (30, 15))
    # A matriz é cimetrica então precisamos apenas das informações acima ou abaixo da diagonal 
    mask = np.tril(np.ones_like(corr_matrix, dtype=bool))
    sns.heatmap(corr_matrix, vmax = .8,mask=mask, square = True, annot = True)
    plt.show()
```

```python
# Matriz de correlação
corrmat = raw_materials.corr()
corrmat_change = raw_materials_change.corr()
plot_heatmap(corrmat)
plot_heatmap(corrmat_change)
```

A matriz de correlação mostra que temos algumas materias-primas com alta correlação no preço, tendo valores acima 0.7 e 0.8. Já a variação percetual tem baixa correlação, com maxima de 0.42. 

### Vamos visualizar melhor a variação ao decorrer do tempo

```python
def plot_columns(df1, df2):
    num_columns = len(df1.columns)
    fig, axs = plt.subplots(num_columns, 2, figsize=(20, num_columns*4))  # 2 coluna lado a lado

    for i in range(num_columns):
        axs[i, 0].plot(df1.index, df1.iloc[:, i], label='Matéria-prima preços')
        axs[i, 0].set_title(df1.columns[i])
        axs[i, 0].legend()
        axs[i, 0].set_xlabel('Anos')
        axs[i, 0].set_ylabel('Preço (dólares por kg)')

        axs[i, 1].plot(df2.index, df2.iloc[:, i], label='Variação percentual', color='orange')
        axs[i, 1].set_title(df2.columns[i])
        axs[i, 1].legend()
        axs[i, 1].set_xlabel('Anos')
        axs[i, 1].set_ylabel('Percentual')

    plt.tight_layout()
    plt.show()
```

```python
plot_columns(raw_materials, raw_materials_change)
```

### Q3: Qual matéria-prima tem a maior e menor variação % de preço em média?

```python
def change_prices(data):
    for column in data.columns:
        # data[column].min(), data[column].max()
        print(f"{column} média = {data[column].mean():.2f}")
change_prices(raw_materials_change)
```

### Q4: Quais matérias-primas tiveram mudanças drásticas no preço?

```python
import random

change_list = raw_materials_change.columns
plt.figure(figsize=(15,12))
for i in range(len(change_list)):
   
    r = random.random()
    b = random.random()
    g = random.random()
    color = (r, g, b)
    plt.subplot(4,3,i+1)
    plt.subplots_adjust( hspace=1 ,wspace=0.5)
    plt.plot(raw_materials_change[change_list[i]],c=color)
    plt.xticks(rotation=90)
    plt.title(change_list[i])
    plt.xlabel('Anos')
    plt.ylabel('% Variação')
```

Observa-se que o maiores picos de variação de preço são na ``Madeira serra macia`` e no ``couro``!
